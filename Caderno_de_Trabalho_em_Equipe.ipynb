```python
idx = 0
dic = []

#retornando com os dados do id repassado
try:
  response = session.get(url_teste, params=parameters)
  retorno = json.loads(response.text)

  for idx in range(len(retorno)):
    dic.append(json.loads(response.text))
except response.exceptions.ConnectionError as e:
  print(e)
```

```python
conteudo_dict = {}
conteudo_dict_quote = {}
conteudo_dict_plataforma = {}

conteudo_dict = dic[0]['data']

circulating_supply = []
date_added = []
last_updated = []
name = []
num_market_pairs = []
slug = []
symbol = []
total_supply = []

#plataform
id = []
name_plataform = []
slug_plataform = []
symbol_plataform = []
token_address = []

#quote data
fully_diluted_market_cap = []
last_updated_1 = []
market_cap = []
market_cap_dominance = []
percent_change_1h = []
percent_change_24h = []
percent_change_30d = []
percent_change_60d = []
percent_change_7d = []
percent_change_90d = []
price = []
volume_24h = []
volume_change_24h = []
```

```python
idx = 0 
idx2 = 0

for idx in range(len(conteudo_dict)):
  circulating_supply.append(conteudo_dict[idx]['circulating_supply'])
  date_added.append(conteudo_dict[idx]['date_added'])
  last_updated.append(conteudo_dict[idx]['last_updated'])
  name.append(conteudo_dict[idx]['name']) 
  num_market_pairs.append(conteudo_dict[idx]['num_market_pairs'])
  slug.append(str(conteudo_dict[idx]['slug']))
  symbol.append(str(conteudo_dict[idx]['symbol']))
  total_supply.append(str(conteudo_dict[idx]['total_supply']))

  #buscando os dados quando há platform, pois as vezes está vazip
  conteudo_dict_plataforma = conteudo_dict[idx]['platform']
  if(conteudo_dict_plataforma != None):
    id.append(conteudo_dict_plataforma['id'])
    name_plataform.append(conteudo_dict_plataforma['name'])
    slug_plataform.append(conteudo_dict_plataforma['slug'])
    symbol_plataform.append(conteudo_dict_plataforma['symbol'])
    token_address.append(conteudo_dict_plataforma['token_address'])
  else:
    id.append(str(''))
    name_plataform.append(str(''))
    slug_plataform.append(str(''))
    symbol_plataform.append(str(''))
    token_address.append(str(''))

  #buscando dados de quote
  conteudo_dict_quote = conteudo_dict[idx]['quote']['USD']

  for idx2 in range(len(conteudo_dict_quote)):
    fully_diluted_market_cap.append(str(conteudo_dict_quote['fully_diluted_market_cap']))
    last_updated_1.append(str(conteudo_dict_quote['last_updated']))
    market_cap.append(str(conteudo_dict_quote['market_cap']))
    market_cap_dominance.append(str(conteudo_dict_quote['market_cap_dominance']))
    percent_change_1h.append(str(conteudo_dict_quote['percent_change_1h']))
    percent_change_24h.append(str(conteudo_dict_quote['percent_change_24h']))
    percent_change_30d.append(str(conteudo_dict_quote['percent_change_30d']))
    percent_change_60d.append(str(conteudo_dict_quote['percent_change_60d']))
    percent_change_7d.append(str(conteudo_dict_quote['percent_change_7d']))
    percent_change_90d.append(str(conteudo_dict_quote['percent_change_90d']))
    price.append(str(conteudo_dict_quote['price']))
    volume_24h.append(str(conteudo_dict_quote['volume_24h']))
    volume_change_24h.append(str(conteudo_dict_quote['volume_change_24h']))
    
```

```python
df_conmarket = pd.DataFrame(list(zip(circulating_supply,date_added,last_updated,name,num_market_pairs,slug,symbol,total_supply,id,name_plataform,slug_plataform,symbol_plataform,token_address,fully_diluted_market_cap,last_updated_1,market_cap,market_cap_dominance,percent_change_1h,percent_change_24h,percent_change_30d,percent_change_60d,percent_change_7d,percent_change_90d,price,volume_24h,volume_change_24h)), \
columns = ['circulating_supply','date_added','last_updated','name','num_market_pairs','slug','symbol','total_supply','id','name_plataform','slug_plataform','symbol_plataform','token_address','fully_diluted_market_cap','last_updated_1','market_cap','market_cap_dominance','percent_change_1h','percent_change_24h','percent_change_30d','percent_change_60d','percent_change_7d','percent_change_90d','price','volume_24h','volume_change_24h'])
df_conmarket.to_csv('df_conmarket' + '.csv', encoding='utf-8-sig', index=False)
```
